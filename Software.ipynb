```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('2015-building-energy-benchmarking.xlsx')
```

```python
data.isnull().sum()
```

```python
df=data.drop(['SecondLargestPropertyUseType', 'SecondLargestPropertyUseTypeGFA','ThirdLargestPropertyUseType','ThirdLargestPropertyUseTypeGFA',
              'YearsENERGYSTARCertified','ENERGYSTARScore','Comment','Outlier','2010 Census Tracts','City Council Districts'], axis = 1)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='YearBuilt', y='GHGEmissions(MetricTonsCO2e)', hue='PrimaryPropertyType', data=data)
plt.title('Relationship between Year Built and GHG Emissions by Property Type')
plt.xlabel('Year Built')
plt.ylabel('GHG Emissions (Metric Tons CO2e)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='BuildingType', y='GHGEmissions(MetricTonsCO2e)', data=data)
plt.title('Distribution of GHG Emissions by Building Type')
plt.xlabel('Building Type')
plt.ylabel('GHG Emissions (Metric Tons CO2e)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='ComplianceStatus', y='GHGEmissionsIntensity(kgCO2e/ft2)', data=data)
plt.title('GHG Emissions Intensity by Compliance Status')
plt.xlabel('Compliance Status')
plt.ylabel('GHG Emissions Intensity (kgCO2e/ft2)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='CouncilDistrictCode', y='GHGEmissions(MetricTonsCO2e)', data=data, estimator=sum)
plt.title('Total GHG Emissions by Council District')
plt.xlabel('Council District Code')
plt.ylabel('Total GHG Emissions (Metric Tons CO2e)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='BuildingType', y='GHGEmissions(MetricTonsCO2e)', hue='ComplianceStatus', data=data, estimator=sum)
plt.title('GHG Emissions by Building Type and Compliance Status')
plt.xlabel('Building Type')
plt.ylabel('Total GHG Emissions (Metric Tons CO2e)')
plt.legend(title='Compliance Status')
plt.xticks(rotation=45)
plt.show()
```

```python
heatmap_data = data.pivot_table(index='YearBuilt', columns='CouncilDistrictCode', values='GHGEmissionsIntensity(kgCO2e/ft2)')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5)
plt.title('GHG Emissions Intensity by Year Built and Council District Code')
plt.xlabel('Council District Code')
plt.ylabel('Year Built')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Neighborhood', y='GHGEmissions(MetricTonsCO2e)', hue='BuildingType', data=data)
plt.title('GHG Emissions by Neighborhood and Building Type')
plt.xlabel('Neighborhood')
plt.ylabel('GHG Emissions (Metric Tons CO2e)')
plt.xticks(rotation=45)
plt.legend(title='Building Type')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='PrimaryPropertyType', y='GHGEmissions(MetricTonsCO2e)', hue='ComplianceStatus', data=data, estimator=sum, ci=None)
plt.title('Average GHG Emissions by Property Type and Compliance Status, grouped by Building Type')
plt.xlabel('Primary Property Type')
plt.ylabel('Total GHG Emissions (Metric Tons CO2e)')
plt.xticks(rotation=45)
plt.legend(title='Compliance Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
g = sns.histplot(data, x='GHGEmissionsIntensity(kgCO2e/ft2)', hue='ComplianceStatus', multiple='stack', bins=30, kde=True)
g.set_title('Distribution of GHG Emissions Intensity by Council District Code')
g.set_xlabel('GHG Emissions Intensity (kgCO2e/ft2)')
g.set_ylabel('Frequency')
plt.legend(title='Compliance Status')
plt.show()
```

```python
# Analyze the relationship between building age and energy efficiency
df['BuildingAge'] = 2015 - df['YearBuilt']
plt.figure(figsize=(12, 8))
sns.scatterplot(x='BuildingAge', y='SiteEUI(kBtu/sf)', hue='PrimaryPropertyType', data=df)
plt.title('Relationship between Building Age and Energy Efficiency')
plt.xlabel('Building Age (years)')
plt.ylabel('Site Energy Use Intensity (kBtu/sf)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Compare energy efficiency of different property types
plt.figure(figsize=(14, 8))
sns.boxplot(x='PrimaryPropertyType', y='SiteEUI(kBtu/sf)', data=df)
plt.title('Energy Efficiency by Property Type')
plt.xlabel('Primary Property Type')
plt.ylabel('Site Energy Use Intensity (kBtu/sf)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze the impact of building size on energy consumption
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PropertyGFATotal', y='SiteEnergyUse(kBtu)', hue='PrimaryPropertyType', data=df)
plt.title('Relationship between Building Size and Energy Consumption')
plt.xlabel('Total Gross Floor Area (sq ft)')
plt.ylabel('Site Energy Use (kBtu)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Analyze the trend of energy efficiency over time
df['YearBuilt'] = pd.to_datetime(df['YearBuilt'], format='%Y')
yearly_efficiency = df.groupby('YearBuilt')['SiteEUI(kBtu/sf)'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='YearBuilt', y='SiteEUI(kBtu/sf)', data=yearly_efficiency)
plt.title('Trend of Energy Efficiency Over Time')
plt.xlabel('Year Built')
plt.ylabel('Average Site Energy Use Intensity (kBtu/sf)')
plt.show()
```
